# Drop column

```python
confirmed_df=confirmed_df.drop(columns=['Lat','Long','Province/State'])
# confirmed_df=confirmed_df.drop(columns=['Province/State'])
deaths_df=deaths_df.drop(columns=['Lat','Long','Province/State'])
recovered_df=recovered_df.drop(columns=['Lat','Long','Province/State'])
lats_df=lats_df.drop(columns=['Long','Province/State','Country/Region'])
longs_df=longs_df.drop(columns=['Lat','Country/Region','Province/State'])
Ps_df=Ps_df.drop(columns=['Lat','Long','Country/Region'])
```

# Group by country

```python
confirmed_df=confirmed_df.groupby(by='Country/Region').aggregate(np.sum).T
confirmed_df.index.name='Date'
confirmed_df = confirmed_df.reset_index()
deaths_df=deaths_df.groupby(by='Country/Region').aggregate(np.sum).T
deaths_df.index.name='Date'
deaths_df = deaths_df.reset_index()
recovered_df=recovered_df.groupby(by='Country/Region').aggregate(np.sum).T
recovered_df.index.name='Date'
recovered_df = recovered_df.reset_index()
lats_df=lats_df.groupby(by='Lat').aggregate(np.sum).T
lats_df.index.name='Date'
lats_df=lats_df.reset_index()

longs_df=longs_df.groupby(by='Long').aggregate(np.sum).T
longs_df.index.name='Date'
longs_df=longs_df.reset_index()

Ps_df=Ps_df.groupby(by='Province/State').aggregate(np.sum).T
Ps_df.index.name='Date'
Ps_df=Ps_df.reset_index()
```

# Melt

```python
confirmed_melt_df=confirmed_df.melt(id_vars='Date').copy()
confirmed_melt_df.rename(columns={'value':'Confirmed'}, inplace=True)
deaths_melt_df=deaths_df.melt(id_vars='Date').copy()
deaths_melt_df.rename(columns={'value':'Deaths'}, inplace=True)
recovered_melt_df=recovered_df.melt(id_vars='Date').copy()
recovered_melt_df.rename(columns={'value':'Recovered'}, inplace=True)
lats_melt_df=lats_df.melt(id_vars='Date').copy()
lats_melt_df.rename(columns={'value':'Lats'},inplace=True)
longs_melt_df=longs_df.melt(id_vars='Date').copy()
longs_melt_df.rename(columns={'value':'Longs'}, inplace=True)
Ps_melt_df=Ps_df.melt(id_vars='Date').copy()
Ps_melt_df.rename(columns={'value': 'Province/States'},inplace=True)
```

# Date format

```python
confirmed_melt_df['Date']=pd.to_datetime(confirmed_melt_df['Date'])
confirmed_melt_df['Date']=confirmed_melt_df['Date'].dt.strftime("%m/%d/%y")
deaths_melt_df['Date']=pd.to_datetime(deaths_melt_df['Date'])
deaths_melt_df['Date']=deaths_melt_df['Date'].dt.strftime("%m/%d/%y")
recovered_melt_df['Date']=pd.to_datetime(recovered_melt_df['Date'])
recovered_melt_df['Date']=recovered_melt_df['Date'].dt.strftime("%m/%d/%y")
```

```python
max_date=confirmed_melt_df['Date'].max()
max_date
```

# Main KPI

```python
fig= go.Figure()
fig.add_trace(go.Indicator(
            mode='number',value=(int)(total_confirmed),
            number={'valueformat':'0,f'},
            title={'text':'Total Confirmed Case'},
            domain={'row':0, 'column':0}
))
fig.add_trace(go.Indicator(
            mode='number',value=(int)(total_deaths),
            number={'valueformat':'0,f'},
            title={'text':'Total Deaths Case'},
            domain={'row':0, 'column':1}
))
fig.add_trace(go.Indicator(
            mode='number',value=(int)(total_recovered),
            number={'valueformat':'0,f'},
            title={'text':'Total recovered Case'},
            domain={'row':1, 'column':0}
))
fig.add_trace(go.Indicator(
            mode='number',value=(int)(total_active),
            number={'valueformat':'0,f'},
            title={'text':'Total Active Case'},
            domain={'row':1, 'column':1}
))
fig.update_layout(grid={'rows':2,'columns':2, 'pattern':'independent'})
fig.show()
```
